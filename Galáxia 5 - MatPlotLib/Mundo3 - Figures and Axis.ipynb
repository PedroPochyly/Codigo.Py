```python
import yfinance as yf
import matplotlib.pyplot as plt

acao = input("Digite o Ticker de uma empresa: ").upper().strip() + ".SA"

try:
    df = yf.download(acao)["Close"]

    if df.empty:
        print("⚠️ Nenhum dado encontrado. Verifique se o ticker está correto ou se a ação foi delistada.")
    else:
        fig, ax = plt.subplots(figsize=(10, 5))
        df.plot(ax=ax)
        ax.set_title(f"Preço de Fechamento - {acao}")
        ax.set_ylabel("Preço (R$)")
        ax.set_xlabel("Data")
        ax.grid(True)
        plt.show()

except Exception as e:
    print(f"Erro ao buscar dados: {e}")
```
